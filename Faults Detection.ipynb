## Now splitting data in training and test data

```python
from sklearn.model_selection import train_test_split
```

```python
Attributes_scaled
```

```python
faults
```

```python
Target = []
for i in range(faults.shape[0]):
    if df['Pastry'].values[i]==1:
        Target.append('Pastry')
    elif df['Z_Scratch'].values[i]==1:
        Target.append('Z_Scratch')
    elif df['K_Scatch'].values[i]==1:
        Target.append('K_Scatch')
    elif df['Stains'].values[i]==1:
        Target.append('Stains')
    elif df['Dirtiness'].values[i]==1:
        Target.append('Dirtiness')
    elif df['Bumps'].values[i]==1:
        Target.append('Bumps')
    else:
        Target.append('Other_Faults')
        
Target = pd.DataFrame(Target)
df['Target']=Target
```

```python
x_train,x_test,y_train,y_test = train_test_split(Attributes_scaled,Target,test_size=0.3,random_state=4)
```

```python
len(x_train)
```

```python
len(x_test)
```

# Applying SVM

```python
from sklearn import svm
model = svm.SVC(C=1,gamma="auto",kernel="linear")
```

```python
print(model.fit(x_train,y_train))
```

```python
print(model.score(x_test,y_test))
```

## Accuracy is 73%

```python
y_pred = model.predict(x_test)
```

```python
print(y_pred)
```

```python
print(y_train)
```

## Now we will built confusion matrix to analyse our prediction

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test,y_pred)
print(cm)
```

## For better visualisation let's built a heatmap

```python
plt.figure(figsize=(10,7))
sn.heatmap(cm,annot=True)
plt.xlabel("predicted")
plt.ylabel("Actual")
plt.show()
```
